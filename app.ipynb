```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('data_user.csv')
df.head()
```

```python
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
map_zoom = 8
my_map = folium.Map(location=map_center, zoom_start=map_zoom)
```

```python
for index, row in df.iterrows():
    if row['carbon_footprint'] > 45:
        marker_color = 'red'
    else:
        marker_color = 'green'
    folium.CircleMarker(location=[row['Latitude'], row['Longitude']],popup=row['carbon_footprint'], radius=10, 
                        color=marker_color, fill=True, fill_opacity=0.7).add_to(my_map)
my_map
```

```python
my_map.save("my_map.html")
```
